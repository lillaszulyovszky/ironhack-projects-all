### respectful scraping 

```python
from time import sleep 
```

```python
#simple example 
for i in range(5):
    print(i)
    sleep(3)
```

```python
#enhance with random integers 
from random import randint
for i in range(5):
    print(i)
    wait_time=randint(1,4)
    print("i will sleep for "+ str(wait_time) +"seconds")
    sleep(wait_time)
```

### putting everything together 

```python
#we are building everything from scratch here 

pages= []

for i in iterations: 
    #takes the iteration, turns it into a string (not needed)
    start_at = str(i)
    #define the url
    url= f"https://www.imdb.com/search/title/?title_type=feature&release_date=1990-01-01,1992-12-31&user_rating=7.5,&start={start_at}&ref_=adv_nxt"
    # download the url content 
    response = requests.get(url)
    #print status code 
    print("Status code: "+ str(response.status_code))
    #to capture the data and fill in the list
    pages.append(response)
    #add a respectful nap 
    wait_time=randint(1,4)
    sleep(wait_time)
```

```python
titles = []
synopsis = []
pages_parsed = []
for i in range(len(pages)):
    # a soup for every search result page
    pages_parsed.append(BeautifulSoup(pages[i].content, "html.parser"))
    # movies:
    movies_html = pages_parsed[i].select("div.lister-item-content")
    
    #loop through the html content of the movies of the current page i 
    
    for j in range(len(movies_html)):
        #extract and appent titles to title list
        titles.append(movies_html[j].select("h3 >a")[0].get_text())
        #extract and appent synopsis to synopsis list
        synopsis.append(movies_html[j].select("p:nth-child(4)")[0].get_text())
        
titles
```

```python
import pandas as pd
movies = pd.DataFrame({'titles':titles,'synopsis':synopsis})
```
