## the hot 100

```python
from bs4 import BeautifulSoup
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# 1. find url and store it in avariable
url = "https://www.billboard.com/charts/hot-100"

# 2. download html with a get request
response = requests.get(url)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup.select('li.chart-list__element:nth-child(1) > button:nth-child(1) > span:nth-child(2) > span:nth-child(1)')
```
