### respectful scraping 

```python
from time import sleep 
```

```python
#simple example 
for i in range(5):
    print(i)
    sleep(3)
```

```python
#enhance with random integers 
from random import randint
for i in range(5):
    print(i)
    wait_time=randint(1,4)
    print("i will sleep for "+ str(wait_time) +"seconds")
    sleep(wait_time)
```

### putting everything together 

```python
#we are building everything from scratch here 

pages= []

for i in iterations: 
    #takes the iteration, turns it into a string (not needed)
    start_at = str(i)
    #define the url
    url_lst=[f"https://www.listchallenges.com/the-top-1000-billboard-hot-100-songs-of-all-time/list/{i}" for i in iterations]    # download the url content 
    response = requests.get(url)
    #print status code 
    print("Status code: "+ str(response.status_code))
    #to capture the data and fill in the list
    pages.append(response)
    #add a respectful nap 
    wait_time=randint(1,4)
    sleep(wait_time)
```
